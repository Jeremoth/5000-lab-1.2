```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

# Create a simple plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label="sin(x)")
plt.title("Example Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```
